```python
import json

import numpy as np
import pandas as pd
import seaborn as sns

from src.chains.hazi_hinam import HaziHinam
from src.chains.osher_ad import OsherAd
from src.chains.rami_levi import RamiLevi
from src.chains.shufersal import Shufersal
from src.chains.victory import Victory
from src.chains.yeinot_bitan import YeinotBitan
from src.chains.yohananof import Yohananof

MONITORED_STORES = {
    repr(Shufersal): [245, 129],
    repr(RamiLevi): [13, 1],
    repr(OsherAd): [3, 1],
    repr(YeinotBitan): [67, 4],
    repr(Yohananof): [20, 1],
    repr(Victory): [89, 1],
    repr(HaziHinam): [1, 6]
}
```

```python
d_s = []
with open('../dbexport', 'r') as fIn:
    for line in fIn:
        try:
            item_d = json.loads(line)
            for date, date_d in item_d.items():
                print(date)
                if date != "_id":
                    for chain, chain_d in date_d.items():
                        print(chain)
                        for store_id, d in chain_d.items():
                            print(store_id)
                            d_s.append({**d, "itemcode": item_d["_id"], "chain": chain, "store_id": store_id})
        except Exception as e:
            print(e)
            continue
            # print(cur_d["_id"])
        # d_s.append(json.loads(line))
```

```python
start_date = '2022-10-12'
end_date = '2022-10-25'
chains = (MONITORED_STORES.keys())
dates_range = pd.date_range(start_date, end_date)
prices = np.zeros((len(dates_range), len(chains)))
item_d = next(d for d in d_s if d.get(start_date) and len(d[start_date].keys()) == len(chains))
print(f"{json.dumps(item_d[start_date], sort_keys=True, indent=4)}")

for i, date in enumerate(dates_range):
    cur_date = str(date.date())
    for j, chain in enumerate(chains):
        # for store_id in MONITORED_STORES[chain]:
        cur_store_id = str(MONITORED_STORES[chain][0])
        print(f"{chain=}")
        # if item_d.get(cur_date, {}).get(chain) is None:
        #     continue
        prices[i][j] = item_d[cur_date][chain][cur_store_id]['final_price']


print(prices)
```
